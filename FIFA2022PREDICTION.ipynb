```python
amdf.rename(columns = {'home_team':'Team1'}, inplace = True)
amdf.rename(columns = {'away_team':'Team2'}, inplace = True)
```

```python
amdf=amdf.drop('tournament',axis=1)
amdf=amdf.drop('city',axis=1)
amdf=amdf.drop('country',axis=1)
indexNames1 = amdf[amdf.Team1.isin(list1)==False].index
amdf.drop(indexNames1 , inplace=True)
indexNames1 = amdf[amdf.Team2.isin(list1)==False].index
amdf.drop(indexNames1 , inplace=True)
amdf=amdf.dropna()
amdf['neutral']=(amdf['neutral']==True).astype("int")
conditions = [amdf['home_score'] > amdf['away_score'], amdf['home_score'] < amdf['away_score']]
choices = [1, 2]
amdf['winner'] = np.select(conditions, choices, default=0)
amdf=amdf.drop('home_score',axis=1)
amdf=amdf.drop('away_score',axis=1)
amdf['date']=pd.to_datetime(amdf['date'])
amdf.index = range(amdf.shape[0])
amdf
```

```python
rf1=RandomForestClassifier(n_estimators=150 , min_samples_split=30, random_state=1)
rf1.fit(x_train,y_train)
rf1.predict(fixturesold)
```

```python
logreg1 = LogisticRegression(random_state=50)
logreg1.fit(x_train,y_train)
logreg1.predict(fixturesold)
```

```python
knn1 =  KNeighborsClassifier(n_neighbors=10)
knn1.fit(x_train,y_train)
knn1.predict(fixturesold)
```

```python
boostc1 = XGBClassifier()
boostc1.fit(x_train,y_train)
boostc1.predict(fixturesold)
```

```python
estimator=[('LR',logreg1),('RF', rf1), ('KNN',knn1), ('BoC',boostc1)]
vot_hard = VotingClassifier(estimators = estimator, voting ='hard',weights=[2,3,0.5,1])
vot_hard.fit(x_train, y_train)
Y_prediction = vot_hard.predict(x_test)
print("\nAccuracy of Voting Classifier", accuracy_score(y_test, Y_prediction))
Y_prediction = vot_hard.predict(fixturesold)
print(Y_prediction)
vot_soft = VotingClassifier(estimators = estimator, voting ='soft',weights=[2,3,0.5,1])
vot_soft.fit(x_train, y_train)
Y_prediction1 = vot_soft.predict(fixturesold)
print(Y_prediction1)
print("\nSimilarity in HardVote and SoftVote", accuracy_score(Y_prediction, Y_prediction1))
predictionOld= vot_soft.predict_proba(fixturesold)
```
